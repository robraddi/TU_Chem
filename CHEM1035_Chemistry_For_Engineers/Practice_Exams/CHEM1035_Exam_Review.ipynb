14. How many atoms are contained in $5.99 × 10^{−14}$ moles of glucose ($C_{6}H_{12}O_{6}$)?
  
a.  $2.42 × 10^{38}$
  
b.  $8.66 × 10^{11}$
  
c.  $1.84 × 10^{21}$
  
d.  $3.61 × 10^{10}$

-------------------------------------------------------------------------

How many **atoms** in $C_{6}H_{12}O_{6}$.

6 carbon, 12 hydrogen, 6 oxygen in 1 mol of $C_{6}H_{12}O_{6}$

6+12+6 = 24 atoms in 1 molecule of glucose

$$(\frac{24 atoms}{\text{1 molecule glucose}})(\frac{6.022*10^{23.}\text{ molecules of glucose}}{\text{1 mol glucose}})(\frac{5.99*10^{-14.} mol}{}) = 8.66*10^{11}\text{ atoms in glucose}$$
