1) $KNO_{3}$, is found in fertilizers, tree stump removal, rocket propellants and fireworks. 


a) What is the IUPAC name for $KNO_{3}$.


Answer: Potassium nitrate 


b) Write an equation for the dissociation of $KNO_{3}$. Example: $ NaClO_{4} \rightarrow Na^{+} + {ClO_{4}}^{-} $


Answer: $KNO_{3} \rightarrow K^{+} + {NO_{3}}^{-}$


c) How many moles of potassium ions would be produced if all 10 g of $KNO_{3}$ dissociated?

Answer:

$$(\frac{\text{1 mol $K^{+}$}}{\text{1 mol $KNO_{3}$}})(\frac{\text{1 mol $KNO_{3}$}}{\text{101.1 g $KNO_{3}$}})(\frac{\text{10 g $KNO_{3}$}}{}) = \text{0.10 mol of $K^{+}$ }$$
