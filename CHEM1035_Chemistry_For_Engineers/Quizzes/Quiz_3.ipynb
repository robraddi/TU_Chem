2) What does $ \frac{P_{1}}{T_{1}} = \frac{P_{2}}{T_{2}}$ imply?


<br>

$$\frac{P_{1}}{T_{1}} = \frac{P_{2}}{T_{2}} \implies  \frac{n_{1}R}{V_{1}} =  \frac{n_{2}R}{V_{2}} $$

The volume and moles of gas is the same before and after. Hence, these are constant. 

<br>




3) Write out the dilution equation. What does it imply?


<br>

$$ M_{1}V_{1} = M_{2}V_{2} $$

$$ \frac{n_{1}}{V_{1}}V_{1} = \frac{n_{2}}{V_{2}}V_{2} \implies n_{1} = n_{2}$$


Regardless of the amount of solvent used there exists the same amount of solute before and after the dilution, such that $n_{1} = n_{2}$.
