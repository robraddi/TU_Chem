**2. Write the Fundamental Equation of Chemical Thermodynamics (FECT).**



$$ \Delta G = \Delta H - T\Delta S $$


Recall, when the reaction is at equilibrium $\Delta G = 0$, which means that if we want to determine the temperature above equilibrium, then we use the following inequality $ 0 > \Delta H - T\Delta S \implies T > \frac{\Delta H}{\Delta S} $


|  $$Reactants \rightarrow Products$$  | $$\Delta G$$ | $$\Delta H$$ | $$T$$ | $$\Delta S$$ | 
| :--: | :--: | :--: | :--: | :--: |
| Spontaneous    | - | - | $\forall \text{ T}$ | + |
| Nonspontaneous | + | + | $\forall \text{ T}$ | - |
| Spontaneous    | - | - | $T < (\Delta H / \Delta S)$ | - |
| Nonspontaneous | + | + | $T > (\Delta H / \Delta S)$ | + |

**For a 0th order integrated rate law:**


$$ \int_{A_{0}}^{A_{t}}\frac{dA}{A^{0}} = \int_{0}^{t} -k  {dt} \implies \int_{A_{0}}^{A_{t}}dA = \int_{0}^{t} -k  {dt}$$

$$ A_{t}-A_{0} = -kt $$


\begin{align}
\text{Integrated rate law: }& A_{t} = -kt + A_{0} \\
\text{Units of k: }& {\displaystyle \ Ms^{-1}}
\end{align}

**For a 1st order integrated rate law:**


$$ \int_{A_{0}}^{A_{t}}\frac{dA}{A^{1}} = \int_{0}^{t} -k  {dt} \implies \int_{A_{0}}^{A_{t}}\frac{1}{A}dA = \int_{0}^{t} -k  {dt}$$

$$ ln(\frac{A_{t}}{A_{0}}) = ln(A_{t}) - ln(A_{0}) = -kt $$



\begin{align}
\text{Integrated rate law: }& ln(A_{t}) = -kt + ln(A_{0})\\
\text{Units of k: }& {\displaystyle \ s^{-1}}
\end{align}


**For a 2nd order integrated rate law:**



$$ \int_{A_{0}}^{A_{t}}\frac{dA}{A^{2}} = \int_{0}^{t} -k  {dt} \implies \int_{A_{0}}^{A_{t}}\frac{1}{A}dA = \int_{0}^{t} -k  {dt}$$

$$ (\frac{1}{A_{t}}-\frac{1}{A_{0}}) = -kt $$


\begin{align}
\text{Integrated rate law: }& \frac{1}{A_{t}} = -kt+\frac{1}{A_{0}}\\
\text{Units of k: }& {\displaystyle \ M^{-1}s^{-1}}
\end{align}

**9. Using the steady state approximation, derive the rate law for the decomposition of ozone in the reaction $2 O_{3} (g) \rightarrow 3 O_{2} (g)$, on the basis of the following intermediate reactions:**

\begin{align}
 O_{3} &\overset{k_{a}}{\rightarrow} O_{2} + O \\
 O_{2} + O &\overset{k_{a}\prime}{\rightarrow} O_{3}  \\
 O + O_{3}  &\overset{k_{b}}{\rightarrow} 2O_{2}
\end{align}

<hr>

<span style="font-size:14px">
    
Ozone, $O_{3}$ is formed at altitudes of 20 to 40 km by UV-radiation ($h\nu$) on $O_{2}$, where $O_{2} \underset{}{\stackrel{h\nu}{\rightarrow}} 2O \hspace{0.25cm}and \hspace{0.25cm} O + O_{2} \rightarrow O_{3}$


\begin{align}
\frac{d[O_{2}]}{dt} &= k_{a} [O_{3}] - k_{a}\prime  [O_{2}]  [O] + k_{b}  [O] [O_{3}] \tag{1} \\
\frac{d[O_{3}]}{dt} &= -k_{a} [O_{3}] + k_{a}\prime  [O_{2}]  [O] + k_{b}  [O] [O_{3}] \tag{2} \\
\frac{d[O]}{dt}     &= k_{a} [O_{3}] - k_{a}\prime  [O_{2}]  [O] - k_{b}  [O] [O_{3}] \tag{3} \\
\end{align}

Perform steady state approximation on $d[O]/dt$


\begin{align}
\frac{d[O]}{dt} = 0 &= k_{a} [O_{3}] - k_{a}\prime  [O_{2}]  [O] - k_{b}  [O] [O_{3}] \\
                  0 &= k_{a} [O_{3}] - [O] (k_{a}\prime  [O_{2}] - k_{b}[O_{3}]) \\
[O] (k_{a}\prime  [O_{2}] - k_{b}[O_{3}]) &= k_{a} [O_{3}] \\
               [O]  &= \frac{k_{a} [O_{3}]}{k_{a}\prime  [O_{2}] - k_{b}[O_{3}]} \\
\end{align}

Now, substituting $[O]$ into equations (1) and (2)


\begin{align}
\frac{d[O_{2}]}{dt} &= k_{a} [O_{3}] - k_{a}\prime  [O_{2}]  [O] + k_{b}  [O] [O_{3}] \\
\frac{d[O_{2}]}{dt} &= k_{a} [O_{3}] - k_{a}\prime  [O_{2}]  (\frac{k_{a} [O_{3}]}{k_{a}\prime  [O_{2}] - k_{b}[O_{3}]}) + k_{b}  (\frac{k_{a} [O_{3}]}{k_{a}\prime  [O_{2}] - k_{b}[O_{3}]}) [O_{3}] \\
\frac{d[O_{2}]}{dt} &= k_{a} [O_{3}] (1 - \frac{k_{a}\prime  [O_{2}]}{k_{a}\prime  [O_{2}] - k_{b}[O_{3}]}) +   (\frac{k_{b}[O_{3}]}{k_{a}\prime  [O_{2}] - k_{b}[O_{3}]})  \\
\frac{d[O_{2}]}{dt} &= k_{a} [O_{3}] (\frac{2k_{b}[O_{3}]}{k_{a}\prime  [O_{2}] - k_{b}[O_{3}]}) 
\end{align}

And we know that $\frac{d[O_{3}]}{dt} = -\frac{d[O_{2}]}{dt}$, so


$$\frac{d[O_{3}]}{dt} = - k_{a} [O_{3}] (\frac{2k_{b}[O_{3}]}{k_{a}\prime  [O_{2}] - k_{b}[O_{3}]}) = \boxed{\frac{(2k_{b}-k_{a})[O_{3}]}{k_{a}\prime  [O_{2}] - k_{b}[O_{3}]}}$$
