**2. a) Wave anatomy: Write down the mathematical form describing a sinusoidal wave $f(x, t)$, annotated with the wavelength $Î»$, phase velocity $v$, and amplitude $A$.**



\begin{align*}
f(x,t=0) &= Acos(\frac{2\pi}{\lambda}(x-\nu t) + \phi) \\

**Do you recall how functions are manipulated?**

```python
%matplotlib inline
import numpy as np
import plot as p

def function(x, t, A=1, nu=1, lam=1, phi=0):
    f = np.array([A*np.cos(2*np.pi/lam*(x[i]-nu*t)+phi) for i in range(len(x))])
    return f
```

```python
x = np.linspace(-100,100,100)
```

**2. b) Plot $f(x, t=0)$ with phase shift of $0$,$\frac{\pi}{4}$,$\frac{\pi}{2}$ and $\frac{3\pi}{4}$**

```python
f = function(x, t=0, A=1, nu=1, lam=1, phi=0)
p.simple_plot(x,f,xlabel='$x$',ylabel='f',Type='line',color='k',fig_size=(8,4))
```

\begin{align*}
f(x,t=0) &= Acos(\frac{2\pi}{\lambda}(x-\nu t) + \phi) \\
&=  Acos(\frac{2\pi}{\lambda}x + \phi) 
\end{align*}

```python
f = function(x, t=0, A=1, nu=1, lam=1, phi=np.pi/4)
p.simple_plot(x,f,xlabel='$x$',ylabel='f',Type='line',color='b',fig_size=(8,4))
```
