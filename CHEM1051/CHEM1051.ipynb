##### Dimensional Analysis:

<u>The big idea:</u>

1. Start with the units your answer should contain i.e., start with what you are looking for.

**If the question asks "...what mass of $H_{2}O$...", then mass should be on top.**


2. Cancel units on top and bottom that match leaving you with the units of your answer.

**If the question says "You have 4.00 mL of $H_{2}O$..."**

$$(\frac{1.00g \text{ of $H_{2}O$}}{mL \text{ of $H_{2}O$}})(4.00mL \text{ of $H_{2}O$})  = 4.00g\text{ of $H_{2}O$}$$

<b>5) The primary component of nail polish remover is acetone.</b>

<img title="www.sigmaaldrich.com" src="https://www.sigmaaldrich.com/content/dam/sigma-aldrich/structure2/194/mfcd00008765.eps/_jcr_content/renditions/mfcd00008765-large.png " width="200" height="200" align="left" > 


Chemical Compound: Acetone

Formula: $C_{3}H_{6}O$

Boiling point: 132.8°F (56°C)

Molar mass: 58.08 g/mol

Density: 784 $kg/m^{3}$

Melting point: -139°F (-95°C)



<b>a. What is the mass, in grams, of 28.56 mL of acetone?</b>

    
Density, $\rho$ is an [intensive property](https://en.wikipedia.org/wiki/Intensive_and_extensive_properties) (independent of the quantity of matter present), and is defined as mass over volume, which are both [extensive properties](https://en.wikipedia.org/wiki/Intensive_and_extensive_properties) (dependent of the quantity):
$$ \rho=\frac{m}{V} $$


$$(\frac{1000g}{1kg})(\frac{784 \text{ kg of $C_{3}H_{6}O$}}{m^{3} \text{ of $C_{3}H_{6}O$}})(\frac{m}{\text 100 cm})^{3}(\frac{\text{28.56 mL of $C_{3}H_{6}O$}}{})  = \text{224.4 g of $C_{3}H_{6}O$}$$

**Make special note that we had to cube the ratio (m/100cm) in order to cancel $m^{3}$ of $C_{3}H_{6}O$.**

<b>b. What is the volume, in mL, of 6.54 g of acetone?</b>

$$(\frac{mL}{cm^3})(\frac{\text 100 cm}{m})^{3}(\frac{m^{3}\text{ of $C_{3}H_{6}O$}}{784 \text{ kg of $C_{3}H_{6}O$}})(\frac{\text{1 kg}}{\text{1000 g}})(\frac{6.54 g}{})  = \text{8.32 mL of $C_{3}H_{6}O$}$$

**Make sure you check significant figures...**

```python
x = (100.**3.)*6.54
y = (784.*1000.)
x/y
```

<b>8)	Silver ions ($Ag^{+}$) react with chloride ions ($Cl^{-}$) to create an insoluble solid, silver chloride (AgCl). How many moles of AgCl would form if 50 mL of a 5 mol/L $Ag^{+}$ solution were mixed with 100 mL of a 2.5 mol/L $Cl^{-}$ solution?</b>

$$Ag^{+} + Cl^{-} \rightarrow AgCl $$

For $Ag^{+}$:

$$(\frac{\text{5 mol}}{L})(\frac{L}{1000 mL})(\frac{\text{50 mL}}{}) = \text{X mol = 0.25 mol}$$

For $Cl^{-}$:

$$(\frac{\text{2.5 mol}}{L})(\frac{L}{1000 mL})(\frac{\text{100 mL}}{}) = \text{Y mol = 0.25 mol}$$

**Note that this problem is similar to <u>determining the Limiting Reagent</u>:**

If X mol < Y mol, then the # of mol AgCl = X mol

If X mol > Y mol, then the # of mol AgCl = Y mol

If X mol = Y mol, then the # of mol AgCl = X mol = Y mol

<b>20) All matter has a property called a specific heat capacity. For silver, this specific heat capacity is 0.24 J/(g*K). If 48 kJ of energy is required to raise the temperature of the silver from 25 °C to 30.4 °C, what is the mass of the silver? (Hint: look at the dimensions of specific heat capacity, then solve)</b>

$$ C_{s} = \frac{q}{m\Delta T} \implies m = \frac{q}{C_{s}\Delta T}$$

$$ \Delta T = 30.4 °C - 25°C = 5.4 K$$

$$m = (\frac{\text{1 kg}}{\text{1000 g}})(\frac{\text{g K}}{\text{0.24 J}})(\frac{\text{1000 J}}{\text{1 kJ}})(\frac{\text{48 kJ}}{\text{}})(\frac{\text{}}{\text{5.4 K}})= \text{37 kg} $$


## Or

$$m = \frac{\text{48 kJ}}{(\frac{kJ}{\text{1000 J}}\frac{\text{0.24 J}}{g K})(\text{5.4 K})} = \text{37 kg} $$

### Alkali Metals: (excludes Hydrogen) lithium (Li), sodium (Na), potassium (K),= rubidium (Rb), caesium (Cs), and francium (Fr)

Properties: shiny, soft, highly reactive metals at standard temperature and pressure and readily lose their outermost electron to form cations with charge +1

Trends: As we move down the column, atomic radii increases. As we move up the column, electronegativity increases

## <span style='color:blue'>Some of the Main Types of Reactions:</span>


##### 1) Combination (synthesis)

$$ \mathrm{A}+\mathrm{B} \rightarrow \mathrm{C} $$

$$ {2H_{2}}_{(g)} + {O_{2}}_{(g)}  \rightarrow  {2H_{2}O}_{(g)}\tag{1} $$

##### 2) Decomposition

$$ \mathrm{C} \rightarrow \mathrm{A}+\mathrm{B}   $$

$$ {2H_{2}O}_{(g)} \rightarrow {2H_{2}}_{(g)} + {O_{2}}_{(g)} \tag{1}$$


##### 3) Single Displacement

$$ \mathrm{AB}+\mathrm{C} \rightarrow \mathrm{CB}+\mathrm{A} $$

$${{CuSO}_{4}}_{(aq)} + Zn_{(s)} \rightarrow {Zn{SO}_{4}}_{(aq)}+Cu_{(s)}\tag{1}$$

##### 4) Double Displacement

$$ \mathrm{AB}+\mathrm{CD} \rightarrow \mathrm{AD}+\mathrm{BC} $$


$$ \mathrm{AgNO}_{3}+\mathrm{NaCl} \rightarrow \mathrm{AgCl}+\mathrm{NaNO}_{3}\tag{1}$$

$$ {NaHCO_{3}}_{(aq)} + {CH_{3}COOH}_{(aq)} \rightarrow {H_{2}CO_{3}}_{(aq)} + {NaCH_{3}COO}_{(s)} \rightarrow {H_{2}O}_{(l)} + {CO_{2}}_{(g)} + {NaCH_{3}COO}_{(s)}\tag{2}$$


##### 5) Neutralization 

Double displacement reactions between strong acids and strong bases results in **salt** and **water**:

$$\mathrm{HCl}_{(aq)}+\mathrm{NaOH}_{(aq)} \rightarrow \mathrm{NaCl}_{(aq)}+\mathrm{H}_{2} \mathrm{O}_{(l)}\tag{1}$$


Double displacement reactions between acids and bases (water as solvent) results in **salt** :

$${NaHCO_{3}}_{(aq)} + {CH_{3}COOH}_{(aq)} \rightarrow {H_{2}CO_{3}}_{(aq)} + {NaCH_{3}COO}_{(s)}\tag{2}$$



##### 6) Combustion

These reactions always result in a change in enthalpy, $\Delta H_{combust} < 0$ and forms $CO_{2}$ and $H_{2}O$.

$$CH_{4} + 2O_{2} \rightleftharpoons CO_{2} + 2H_{2}O\hspace{0.2cm};\hspace{0.2cm} \Delta H < 0 \tag{1}$$

Suppose wood has the chemical formula $\mathbf{C}_{6} \mathbf{H}_{12} \mathbf{O}_{6}$. Then,

$$\mathbf{C}_{6} \mathbf{H}_{12} \mathbf{O}_{6} \quad+6 \mathbf{O}_{2}\rightleftharpoons\quad 6 \mathrm{CO}_{2} \quad+\quad 6 \mathbf{H}_{2} \mathbf{O}\hspace{0.2cm};\hspace{0.2cm} \Delta H <0 \tag{2}$$


##### 7) Redox


$${{CuSO}_{4}}_{(aq)} + Zn_{(s)} \rightarrow {Zn{SO}_{4}}_{(aq)}+Cu_{(s)}\tag{1}$$
