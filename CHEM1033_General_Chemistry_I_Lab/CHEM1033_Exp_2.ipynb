### Part A. Determine Avogadro's Number From the Density of Copper

<h2 align=justify; style='font-size: 12pt;
                  font-family: times;
                  margin-top: 0.0in;

                  '>   
    <p> Copper crystallizes into a face centered cubic (fcc) unit cell. </p>
    <p> The atomic mass of Cu is 63.546 g/mol and the length of one edge in the unit cell is 361.2 pm. </p>
    <p> There is 8 corners and 6 faces of the unit cell, where there exists a Cu atom (each of which are shared with other unit cells) at each of these positions.
<figure>
    <img src="./images/fcc.png" width="350" height="350" align="center" > 
</figure>  
    <br/>
    <p> I. Solve for b </p> 
    <p> $$ b^{2} = l^{2} + l^{2} = 2l^{2}  $$ </p>
    <p> $$ b = \sqrt{2l^{2}} = 4r $$ </p>
    <p> II. Solve for l </p> 
    <p> $$ 2l^{2} = b^{2}  $$ </p>
    <p> $$ l = \sqrt{\frac{b^{2}}{2}} = \frac{b}{\sqrt{2}} = \frac{4r}{\sqrt{2}} = 2\sqrt{2}r  $$ </p>
    <br/><br/><br/><br/>
    <p> 1. Determine the volume, V of the unit cell.</p>  
    <p> $$ V_{cube} = l^{3}, \text{ where l is the edge length of the unit cell.}  $$ </p>  
    <p> 2. Calculate the mass of the unit cell.</p>  
    <p> $$ Density = \rho = \frac{mass}{volume}$$ </p>
    <p> $$ Mass_{Cell} = \rho_{Calc}Volume_{Cell}$$ </p>
    <p> 3. Determine the number of atoms that occupy the unit cell.</p>  
    <p> $$ \text{Atoms/Unit Cell } =  (\frac{1}{8}\times \text{# of Corner Atoms}) + (\frac{1}{2} \times \text{# of Face Centered Atoms} )$$ </p>
    <p> $$ \text{Atoms/Unit Cell } =  (\frac{1}{8}\times 8) + (\frac{1}{2} \times 6 ) = 4 $$ </p>
    <p> 4. Use dimensional analysis to calculate Avogadro's number from the number of Cu atoms per unit cell, the mass per unit cell, and the molar mass of Cu.</p>  
    <p> $$ (\frac{\text{4 Cu Atoms}}{\text{1 Cu Unit Cell}})\times(\frac{\text{1 Cu Unit Cell}}{\text{Mass of Unit Cell}})\times(\frac{63.546 g}{\text{1 mol}}) = \text{# of Cu Atoms} $$ </p>

    <p> 1. Determine the volume of stearic acid/hexane solution, $V_{SAH}$ using your calibration ratio (drops/mL), $\alpha$ and the average number of drops to form the monolayer. </p>  
    <p> $$  V_{SAH} = \alpha^{-1}\times(\text{Avg. Drops}) $$ </p>  
    <p> 2. Calculate the mass of stearic acid in the monolayer, $m_{stearic}$ using the concentration of stearic acid/hexane solution, $M_{SAH}$. 
    <br/>Note: Molar Mass of Stearic acid is 284.48 g/mol.</p>  
    <p> $$ m_{stearic} =  (M_{SAH})\times(V_{SAH}) $$ </p>
    <p> 3. Calculate the volume of stearic acid in the monolayer, $V_{Stearic}$</p>  
    <p> $$ V_{Stearic} = (\rho_{Steric Acid})^{-1}\times m_{stearic} $$</p>
    <p> $$  $$ </p>
    <p> 4. Calculate the thickness of the monolayer, h where the volume of stearic acid in the monolayer can be approximated by $V=\pi r^{2}h$. </p>  
    <p> $$ h = \frac{\pi r^{2}}{V} = \frac{\pi (\frac{d}{2})^{2}}{V} $$ </p>
    <p> 5. Determine the volume of a carbon atom. <br/> Note that h/18 can be used as a good approximation for the diameter of a carbon atom.</p>  
    <p> $$ V_{\text{Carbon atom}}  = \frac{4}{3}\pi r^{3} = \frac{4}{3}\pi \frac{d}{2}^{3} = \frac{4}{3}\pi (\frac{h}{36})^{3} $$ </p>
    <p> 6. Determine Avogadro's number.</p>
    <p> $$ N_{A} = (\frac{\text{1 Carbon Atom}}{V_{\text{Carbon atom}}})\times(\rho_{\text{Carbon}})^{-1}\times(\frac{\text{12.0107 g Carbon}}{\text{1 mol Carbon}}) $$ </p>
